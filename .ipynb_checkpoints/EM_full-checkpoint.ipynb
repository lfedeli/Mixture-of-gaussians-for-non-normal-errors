```python
# Import the needed modules
import numpy as np
import tensorflow as tf
%matplotlib inline
import matplotlib.pyplot as plt
import math
rng = np.random
from sklearn.model_selection import KFold, train_test_split
from sklearn.metrics import r2_score
from sklearn import mixture
import csv
```

## Initial fit

The algorithm is initialized training the data with a preliminary linear gaussian regression. We calculate the residuals and fit with a mixture of gaussians. This operation i

```python
def initialtrain():
    train_X = np.asarray(listax[0] + listax[1] + listax[2] + listax[3] + listax[4])
    train_Y = np.asarray(listay[0] + listay[1] + listay[2] + listay[3] + listay[4])
    numero = train_X.shape[0]
    for epoch in range(training_epochs): 
        for (x, y) in zip(train_X, train_Y):
            sess.run( optimizerg, feed_dict={X: x, Y: y, n_samples: numero })  
        if (epoch+1) % display_step == 0:
                c = sess.run(costg, feed_dict={X: train_X, Y: train_Y, n_samples: numero  })
                print("Epoch:", '%04d' % (epoch+1), "cost=", "{:.9f}".format(c), \
                                      "W=", sess.run(Wg), "b=", sess.run(bg)) 

def initialfit():
    train_X = np.asarray(listax[0] + listax[1] + listax[2] + listax[3] + listax[4])
    train_Y = np.asarray(listay[0] + listay[1] + listay[2] + listay[3] + listay[4])
    Wfing = sess.run( Wg )
    bfing = sess.run(bg)
    ypredg = Wfing*train_X + bfing
    residuals = train_Y - ypredg
    residualsnew = residuals.reshape( (residuals.size, 1) )
    clf = mixture.GaussianMixture(n_components=numbercomp, covariance_type='spherical')
    clf.fit(residualsnew)
    return clf               

def initialize(clf, listax,listay):
    train_X =  np.reshape( np.asarray(listax[0] + listax[1] + listax[2] + listax[3] + listax[4]), (1,nsamples))
    train_Y =  np.reshape( np.asarray(listay[0] + listay[1] + listay[2] + listay[3] + listay[4]), (nsamples,1))
    with tf.variable_scope("rank2", reuse = True):
        alpha = tf.get_variable("alpha", shape = [numbercomp, 1] )  
        mu = tf.get_variable("mu",shape = [numbercomp,1] ) 
        beta  = tf.get_variable("beta", shape = [1,1]) 
        tau   = tf.get_variable("tau",shape = [numbercomp, 1 ] ) 
        weights = tf.get_variable("weights",shape = [numbercomp, 1 ] )
        gamma = tf.get_variable("gamma",shape = [numbercomp + 1, 1 ] ) 
        Y     = tf.get_variable("Y",shape = [nsamples,1 ] )
        X     = tf.get_variable("X",shape = [1,nsamples ] )
        
    a = np.reshape(clf.means_,(numbercomp,))
    b = np.dot(a, clf.weights_)
    clf.means_ = clf.means_ - b   
        
    W = np.reshape( sess.run(Wg), (1,1) )
    binit = np.reshape( sess.run(bg), (1,1) )
    
    medie = np.reshape( clf.means_, (numbercomp, 1) )
    variances = np.reshape( clf.covariances_, (numbercomp, 1) )
    we = np.reshape( clf.weights_, (numbercomp, 1) )
    sess.run( [tf.assign(beta, W), tf.assign(mu,medie), tf.assign(tau,variances), tf.assign(weights,we) ] ) 
    sess.run( [tf.assign(X,train_X), tf.assign(Y,train_Y)] )
    sess.run( [tf.assign(alpha, tf.add(mu,binit) )] )
    sess.run( tf.assign( gamma, tf.concat([alpha,beta],0) ) ) 
    print("verify initial constraint ", sess.run( tf.reduce_sum(tf.multiply(mu,weights ) )) )


def buildXgrande(): 
    train = np.asarray(listax[0] + listax[1] + listax[2] + listax[3] + listax[4])
    Xgrande = np.zeros( (numbercomp,nsamples, numbercomp + 1) )
    for i in range(numbercomp):
        a = np.zeros((1, numbercomp + 1))
        a[0][i] = 1.
        loc = np.tile(a,(nsamples,1))
        for j in range(nsamples):
            loc[j][numbercomp] = train[j]
        Xgrande[i,:,:] = loc
    return Xgrande    

def buildZgrande():
    with tf.variable_scope("rank2exp", reuse = True):
        prob = tf.get_variable("probabilities",shape = [numbercomp, nsamples ] )
    output = sess.run( prob )
    Zgrande = np.zeros( (numbercomp, nsamples, nsamples) )
    for i in range(numbercomp):
        diago = np.diag( output[i][:] )
        Zgrande[i][:][:] = diago
    return Zgrande
```

```python
# Parameters
learning_rate = 0.07
training_epochs =  70
display_step = 50
numbercomp = 3
nsamples = 350
Pi = tf.constant( 3.141592653589793238, dtype = "float")
```

```python
with tf.variable_scope("rank2"):
    tf.get_variable("b_ottimale", shape = [1, 1] )  
    tf.get_variable("alpha", shape = [numbercomp, 1] )  
    tf.get_variable("beta", shape = [1,1]) 
    tf.get_variable("mu",shape = [numbercomp,1] ) 
    tf.get_variable("tau",shape = [numbercomp, 1 ] ) 
    tf.get_variable("gamma",shape = [numbercomp + 1, 1 ] ) 
    tf.get_variable("weights",shape = [numbercomp, 1 ] ) 
    #tf.get_variable("z.h",shape = [numbercomp, 1 ] ) 
    tf.get_variable("Y",shape = [nsamples,1 ] )
    tf.get_variable("X",shape = [1,nsamples ] )

    
with tf.variable_scope("rank2exp"):
    tf.get_variable("alpha", shape = [numbercomp, nsamples] )  
    Y = tf.get_variable("Y",shape = [numbercomp, nsamples ] )
    tf.get_variable("X",shape = [numbercomp ,nsamples ] )
    mu = tf.get_variable("mu",shape = [numbercomp, nsamples] ) 
    tau = tf.get_variable("tau",shape = [numbercomp, nsamples] ) 
    tf.get_variable("weights",shape = [numbercomp, nsamples ] ) 
    tf.get_variable("probabilities",shape = [numbercomp, nsamples ] )
    tf.get_variable("z",shape = [numbercomp, nsamples] ) 
    gaussian = tf.multiply( 1./(tf.sqrt(2.*Pi*tau)) , tf.exp(tf.div(-tf.pow( tf.subtract(Y, mu ), 2.), 2.*tau)))
    
with tf.variable_scope("rank3"):
    tf.get_variable("Y",shape = [numbercomp,  nsamples, 1 ] )
    tf.get_variable("gamma",shape = [numbercomp, numbercomp + 1,1 ] ) 
    tf.get_variable("tau",shape = [numbercomp, 1, 1] ) 
    tf.get_variable("Zh", shape = [numbercomp, nsamples, nsamples ])   

Xh =  tf.placeholder("float", shape = [numbercomp, nsamples, numbercomp + 1 ])  
  
```

### Variables for the optimization problem

```python
atrain = tf.get_variable("atrain",shape=[])
btrain = tf.get_variable("btrain",shape=[])
    
def energia_training(trainsetx,trainsety, muop, tauop, wshop,ns): 
    X_loc = tf.constant(trainsetx, dtype = "float32")
    Y_loc = tf.tile( tf.reshape( tf.constant(trainsety, dtype = "float32"), [1,ns]) , [numbercomp,1])
    betaX_loc = tf.tile( tf.reshape( tf.scalar_mul(atrain,X_loc) , [1,ns]), [numbercomp,1] )
    alpha_loc =     tf.tile( tf.add(muop, btrain), [1,ns] )
    mu_loc = tf.add(alpha_loc, betaX_loc)
    tau_loc = tf.tile( tauop, [1,ns] )
    whs_loc = tf.tile(wshop, [1,ns])
    gaussian_loc = tf.multiply( 1./(tf.sqrt(2.*Pi*tau_loc)) ,\
                               tf.exp(tf.div(-tf.pow( tf.subtract(Y_loc, mu_loc ), 2.), 2.*tau_loc)))
    numP = tf.multiply(whs_loc,gaussian_loc)
    Prob = tf.div( numP, tf.reduce_sum(numP,0 ) )
    
    num = tf.pow( tf.subtract(Y_loc, mu_loc), 2.)
    term = tf.multiply(Prob, tf.div(num,tau_loc) )
    
    reducedP = tf.reshape( tf.reduce_sum(Prob,1), [numbercomp,1])
    lc2 = 0.5*tf.reduce_sum(tf.multiply(reducedP,tf.log( tauop)) )
    
    lc1 = - tf.reduce_sum( tf.multiply( reducedP, tf.log(wshop)  ))
    
    energia = tf.add( tf.add( lc2, tf.reduce_sum( tf.reduce_sum( term) )), lc1)
    return energia
    
   
```

```python
#################################################################for the first step
X = tf.placeholder("float")
Y = tf.placeholder("float")
n_samples = tf.placeholder("float", shape = (), name = 'numeber_of_samples')

Wg = tf.Variable(rng.randn(), name="gaussian_model_weight")
bg = tf.Variable(rng.randn(), name="gaussian_model_bias")

predg = tf.add(tf.multiply(X, Wg), bg)
costg = tf.reduce_sum(tf.pow(predg-Y, 2))/(2*n_samples)
# Gradient descent
optimizerg = tf.train.GradientDescentOptimizer(learning_rate).minimize(costg)
```

```python
def update_probabilities():
    with tf.variable_scope("rank2", reuse = True):
        alphash  =  tf.get_variable("alpha", shape = [numbercomp, 1] )  
        Xsh      =  tf.get_variable("X",shape = [1 ,nsamples ] )
        beta     =  tf.get_variable("beta", shape = [1,1]) 
        taush    =   tf.get_variable("tau",shape = [numbercomp, 1 ] ) 
        wsh      =   tf.get_variable("weights",shape = [numbercomp, 1 ] ) 
        Ysh      =   tf.get_variable("Y",shape = [nsamples,1 ] )
    with tf.variable_scope("rank2exp", reuse= True):
        alpha   =  tf.get_variable("alpha", shape = [numbercomp, nsamples] )  
        X       =  tf.get_variable("X",shape = [numbercomp ,nsamples ] )
        mu      =  tf.get_variable("mu",shape = [numbercomp, nsamples] )
        tau     =  tf.get_variable("tau",shape = [numbercomp, nsamples] ) 
        weights =  tf.get_variable("weights",shape = [numbercomp, nsamples ] ) 
        Y       =  tf.get_variable("Y",shape = [numbercomp, nsamples ] )
        prob    =  tf.get_variable("probabilities",shape = [numbercomp, nsamples ] )
    sess.run( tf.assign( alpha, tf.tile( alphash, (1,nsamples) )  ) )    
    sess.run( tf.assign( X, tf.multiply( beta, tf.tile( Xsh, (numbercomp,1) ) )  ) ) 
    sess.run( tf.assign( mu, tf.add(alpha, X) ) )
    
    sess.run( tf.assign( tau, tf.tile( taush, (1,nsamples) )  ) )
    sess.run( tf.assign( weights, tf.tile( wsh, (1,nsamples) )  ) )
    
    sess.run( tf.assign( Y, tf.tile( tf.transpose(Ysh), (numbercomp,1) )  ) )
    
    num = tf.multiply(weights, gaussian)
    den = tf.reduce_sum(num,0)
    sess.run( tf.assign(prob, tf.div(num,den)  )  )

def update_weights():
    with tf.variable_scope("rank2", reuse = True):
         wsh      =   tf.get_variable("weights",shape = [numbercomp, 1 ] ) 
    with tf.variable_scope("rank2exp", reuse= True):
         prob    =  tf.get_variable("probabilities",shape = [numbercomp, nsamples ] )        
    output = sess.run( tf.reduce_sum( prob,1 )/nsamples )
    output = np.reshape(output,(numbercomp,1))
    sess.run(tf.assign(wsh,output))


    

def update_gamma_alpha_beta():
    with tf.variable_scope("rank2", reuse = True):
        taush = tf.get_variable("tau",shape = [numbercomp, 1 ] ) 
        gamma = tf.get_variable("gamma",shape = [numbercomp + 1, 1 ] ) 
        Y =     tf.get_variable("Y",shape = [nsamples,1 ] )    
        alpha = tf.get_variable("alpha", shape = [numbercomp, 1] )  
        beta  =  tf.get_variable("beta", shape = [1,1]) 
    with tf.variable_scope("rank3", reuse = True):
        tau = tf.get_variable("tau",shape = [numbercomp, 1, 1] ) 
        Zh  =  tf.get_variable("Zh", shape = [numbercomp, nsamples, nsamples ]) 
    sess.run( tf.assign( tau,  tf.expand_dims(taush,2) ) )
    Xgrande = buildXgrande()
    Zgrande = buildZgrande()
    sess.run(tf.assign(Zh, Zgrande ) )
    
    product1 = tf.matmul(  tf.transpose(Xh, perm=[0,2,1]), Zh )
    product = tf.matmul( product1 , Xh) 
    M = tf.reduce_sum( tf.div(product,tau) ,0 )
    N = tf.reduce_sum( tf.div(product1,tau), 0 )   
    gammanew = tf.matmul( tf.matmul( tf.matrix_inverse(M), N ), Y )
    sess.run( tf.assign(gamma,gammanew), feed_dict={Xh: Xgrande}  )
    split0, split1 = tf.split(gamma, [numbercomp, 1], 0) 
    sess.run(tf.assign(alpha,split0))
    sess.run(tf.assign(beta,split1))
    

def update_tau():
    with tf.variable_scope("rank2", reuse = True):
        gammash = tf.get_variable("gamma",shape = [numbercomp + 1, 1 ] ) 
        Ysh      =   tf.get_variable("Y",shape = [nsamples,1 ] )
        tau      =   tf.get_variable("tau",shape = [numbercomp, 1 ] )
    with tf.variable_scope("rank2exp", reuse= True):
        prob  = tf.get_variable("probabilities",shape = [numbercomp, nsamples ] )
    with tf.variable_scope("rank3", reuse = True):
        """Zh GIA COSTRUITA quando ho fatto l update di gamma CAPITO?????"""
        Zh  =  tf.get_variable("Zh", shape = [numbercomp, nsamples, nsamples ]) 
        gamma = tf.get_variable("gamma",shape = [numbercomp, numbercomp + 1,1 ] )
        Y     = tf.get_variable("Y",shape = [numbercomp,  nsamples, 1 ] )
    Xgrande = buildXgrande()
    ynew = tf.tile( tf.expand_dims( Ysh, 0 ), (numbercomp,1,1) )
    sess.run(tf.assign(Y,ynew))
    gammanew = tf.tile( tf.expand_dims( gammash, 0 ), (numbercomp,1,1) )
    sess.run(tf.assign(gamma,gammanew))
    
    term3 = tf.subtract( Y, tf.matmul(Xh,gamma))
    term1 = tf.transpose( term3, perm = [0,2,1] ) 
    prodotto = tf.matmul( tf.matmul( term1, Zh ), term3)
    zdoth = tf.expand_dims( tf.expand_dims( tf.reduce_sum(prob,1), 1), 2)
    taunew = tf.reshape( tf.div(prodotto, zdoth), [numbercomp,1] )
    sess.run(tf.assign(tau,taunew), feed_dict={Xh: Xgrande})

    
def calculate_optimal_param():
    with tf.variable_scope("rank2", reuse = True):
        b = tf.get_variable("b_ottimale", shape = [1, 1] )  
        mu = tf.get_variable("mu",shape = [numbercomp,1] )
        alpha =  tf.get_variable("alpha", shape = [numbercomp, 1] )  
        wsh =    tf.get_variable("weights",shape = [numbercomp, 1 ] ) 
    bop = tf.reshape( tf.reduce_sum( tf.multiply(wsh,alpha)), [1,1])
    mufin = tf.subtract(alpha,bop)
    sess.run([tf.assign(b,bop), tf.assign(mu,mufin)])
    print("verify initial constraint dopo ", sess.run( tf.reduce_sum(tf.multiply(mu,wsh ) )) ) 

    


def calculate_energy():
    with tf.variable_scope("rank2", reuse = True):
        taush    =   tf.get_variable("tau",shape = [numbercomp, 1 ] ) 
        wsh      =   tf.get_variable("weights",shape = [numbercomp, 1 ] ) 
    with tf.variable_scope("rank2exp", reuse= True):
        mu      =  tf.get_variable("mu",shape = [numbercomp, nsamples] )
        tau     =  tf.get_variable("tau",shape = [numbercomp, nsamples] ) 
        Y       =  tf.get_variable("Y",shape = [numbercomp, nsamples ] )
        prob    =  tf.get_variable("probabilities",shape = [numbercomp, nsamples ] )
    zdoth = tf.reshape( tf.reduce_sum( prob,1 ), [numbercomp,1])
    lc1 = tf.reduce_sum( tf.multiply(zdoth, tf.log(wsh)) )
    lc2 = - 0.5*tf.reduce_sum( tf.multiply(zdoth, tf.log(taush)) )
    term =  tf.multiply(prob, tf.div( tf.pow( tf.subtract(Y,mu), 2.), tau ) )
    lc3 = - 0.5*tf.reduce_sum( tf.reduce_sum(term,0))
    energia = tf.reshape(tf.add( tf.add(lc1,lc2), lc3), [])
    return sess.run(energia)

```

### Post-porcessing operation

```python
def define_parameters():
    with tf.variable_scope("rank2", reuse = True): 
        b = tf.get_variable("b_ottimale", shape = [1, 1] )    
        a =  tf.get_variable("beta", shape = [1,1]) 
        mu  =  tf.get_variable("mu",shape = [numbercomp,1] ) 
        tau =   tf.get_variable("tau",shape = [numbercomp, 1 ] ) 
        wsh =   tf.get_variable("weights",shape = [numbercomp, 1 ] ) 
    bop = sess.run(b)
    aop = sess.run(a)
    muop = sess.run(mu)
    tauop = sess.run(tau)
    wshop = sess.run(wsh)
    return tf.constant(bop), tf.constant(aop), tf.constant(muop), tf.constant(tauop), tf.constant(wshop)

def print_results():
    with tf.variable_scope("rank2",reuse = True):
        tau = tf.get_variable("tau",shape = [numbercomp, 1 ] ) 
        weights = tf.get_variable("weights",shape = [numbercomp, 1 ] ) 
    print("varianze", sess.run( [tau]) )
    print("pesi", sess.run( [weights]))
    
def plot_results():
    train_X = np.asarray(listax[0] + listax[1] + listax[2] + listax[3] + listax[4])
    train_Y = np.asarray(listay[0] + listay[1] + listay[2] + listay[3] + listay[4])
    with tf.variable_scope("rank2", reuse = True):
        b = tf.get_variable("b_ottimale", shape = [1, 1] )  
        a = tf.get_variable("beta",shape = [1,1] )
    plt.plot(train_X, train_Y, 'ro', label='Original data')
    aa = tf.reshape(a,[])
    bb = tf.reshape(b,[])
    plt.plot(train_X, sess.run(aa) * train_X + sess.run(bb), label='Fitted line')
    plt.legend()
    plt.show()
    
```

```python
init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)
    initialtrain()
    clf = initialfit()
    initialize(clf,listax,listay)
    for i in range(4):
        update_probabilities()
        update_weights()
        update_gamma_alpha_beta()
        update_tau()
        energia = calculate_energy()
        if i%20 == 0:
            print("Expectation-Maximization likelihood after {}-steps".format(i), energia)
    calculate_optimal_param()        
    bop, aop, muop, tauop, wshop = define_parameters() 
    
    
    score_mixgauss = []
    score_gaussian = []
    parameters_mixgauss = []
    parameters_gaussian = []
    learning_rate = 0.000001
    print("")
    print("I can start to train on each dataset")
    print("")
    for i in range(5):
            train_X2 = np.asarray(listax[i])
            train_Y2 = np.asarray(listay[i])
            train_X1, test_X1, train_Y1, test_Y1 = train_test_split( train_X2, train_Y2, test_size=0.35, random_state=42)
            ns = train_X1.shape[0]
            sess.run(tf.assign(atrain, tf.reshape(aop,[]) ))
            sess.run(tf.assign(btrain, tf.reshape(bop,[]) ))
            energy_T = energia_training(train_X1, train_Y1, muop, tauop, wshop, ns)
            print("I am training the dataset #{}".format(i+1))
            optimizer_T = tf.train.GradientDescentOptimizer(0.003).minimize(energy_T) 
            for i in range(300):
                sess.run(optimizer_T)
                for (x, y) in zip(train_X1, train_Y1):
                    sess.run( optimizerg, feed_dict={X: x, Y: y, n_samples: ns })
                if i % 100 == 0:
                    print("costo mixed gauss = ", sess.run(energy_T), \
                        "   costo tot.  gauss = ", sess.run(costg,  feed_dict={X: train_X1, Y: train_Y1, n_samples: ns }) )
            Wfin, Wfing = sess.run( [atrain,Wg] )
            bfin, bfing = sess.run( [btrain, bg])
            ypred = Wfin*test_X1 + bfin
            ypredg = Wfing*test_X1 + bfing
            parameters_mixgauss.append( (Wfin, bfin) )
            parameters_gaussian.append( (Wfing, bfing) )
            score_mixgauss.append( r2_score(test_Y1, ypred) )
            score_gaussian.append( r2_score(test_Y1, ypredg) )  
    for i in range(5):
        print("")
        print("R^2 score on dataset #{}".format(i + 1), "mix of gaussians student = ", score_mixgauss[i],\
              "   total gaussian = ", score_gaussian[i] )
                    
```
